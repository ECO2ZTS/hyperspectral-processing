## Imports

```python
import pandas as pd
import sys

# needed, if the module is not installed:
sys.path.insert(0, '../')
from hprocessing.PlotUtils import (plotEnviImageWithMask,
                                   plotEnviImageWithRectangles)
from hprocessing.ProcessEnviFile import getEnviFile
from hprocessing.ProcessFullDataset import getMask
```

## Plot

```python
hdr, img = getEnviFile(filepath="../data/testfiles/hyp/Auto017.hdr")
masks = pd.read_csv("../data/testfiles/masks/masks_test.csv", sep="\s+")
mask = getMask(masks, 0)
rectangles = [[30, 40, 10, 20]]
```

```python
plotEnviImageWithMask(
    image=img,
    wavelengths=hdr["Wavelength"],
    rectangles=rectangles,
    mask=mask,
    channel=90,
    save_to_file=False
)
```

```python
plotEnviImageWithRectangles(
    image=img,
    wavelengths=hdr["Wavelength"],
    rectangles=rectangles,
    channel=90,
    save_to_file=False
)
```
